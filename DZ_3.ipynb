##Задача 1.2
Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома Если реновации дома не было, то в новом признаке поставьте 0

```python
def _renovated():
    delta_renovated = []
    for _, row in df.iterrows():
        if row['yr_renovated'] > 0:
            delta_renovated.append(row['yr_renovated'] - row['yr_built'])
        else:
            delta_renovated.append(0)
    return delta_renovated

df['delta_renovated'] = _renovated()
df.head()
```

3.1 Найдите среднюю стоимость домов в зависимости от количества спален

```python
from pandas.core.dtypes.cast import astype_array
df['price'].astype('bool')
```

##Задача 3.2
Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
df.groupby('condition').agg({'price': ['min', 'max', 'mean']})
```

##Задача 3.3
Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df.groupby(['waterfront', 'view']).agg({'view': 'count'})
```

```python
df.pivot_table(index='waterfront', columns = 'view', values='price', aggfunc='count', fill_value=0)
```

##Задача 3.4
Каких домов в зависимости от этажности и количества спален больше?

```python
pd.crosstab(index=df['floors'], columns=df['bedrooms'], margins=True, margins_name='Total')
```

##Задача 3.5
Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
df.groupby(['condition', 'grade']).agg({'price': 'median'}).unstack('grade').fillna(0)
```
